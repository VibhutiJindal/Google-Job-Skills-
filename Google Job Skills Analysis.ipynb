```python
#companies involved for recruitment
data['Company'].value_counts()
```

```python
#visualizing company ratio
plt.figure(figsize=(7,9))
data['Company'].value_counts().plot(kind='bar',color='green')
plt.xlabel("Name of Companies",fontsize=15)
plt.title("Companies involved for recruitment",fontsize=20)
plt.show()
```

```python
#most popular 20 job titles of google
data["Title"].value_counts().sort_values(ascending=False).head(20)
```

```python
#visualizing Most Popular 20 Job Titles of Google
plt.figure(figsize=(19,8))
data['Title'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'pink')
plt.title("Most Popular 20 Job Titles of Google", fontsize = 20)
plt.xlabel('Names of Job Titles', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```python
#Most Popular 20 Job Category of Google
data["Category"].value_counts().sort_values(ascending=False).head(20)
```

```python
#visualizing Most Popular 20 Job Category of Google
plt.figure(figsize=(19,8))
data['Category'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightgreen')
plt.title("Most Popular 20 Job Category of Google", fontsize = 20)
plt.xlabel('Names of Category', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```python
#Most Popular 20 Job locations of Google
data["Location"].value_counts().sort_values(ascending=False).head(20)
```

```python
#visualizing Most Popular 20 Job locations of Google
plt.figure(figsize=(19,8))
data['Location'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightblue')
plt.title("Most Popular 20 Job locations of Google", fontsize = 20)
plt.xlabel('Names of Locations', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
#places with most job
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'white').generate(str(data['Location']))

print(wordcloud)
plt.figure(figsize = (12, 12))
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Places with Most Jobs', fontsize = 30)
plt.show()
```

```python
#most popular responsibilities
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'lightgreen', max_words = 100).generate(str(data['Responsibilities']))

print(wordcloud)
plt.figure(figsize = (12, 12))
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Responsibilities', fontsize = 30)
plt.show()
```

```python
#Most Popular Educational Requirements
wordcloud = WordCloud(background_color = 'lightblue', max_words = 80).generate(str(data['Minimum Qualifications']))

print(wordcloud)
plt.figure(figsize = (12, 12))
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Educational Requirements', fontsize = 30)
plt.show()
```

```python
#Most Popular Preferred Requirements
wordcloud = WordCloud(background_color = 'yellow', max_words = 80).generate(str(data['Preferred Qualifications']))

print(wordcloud)
plt.figure(figsize = (12, 12))
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Preferred Requirements', fontsize = 30)
plt.show()
```
